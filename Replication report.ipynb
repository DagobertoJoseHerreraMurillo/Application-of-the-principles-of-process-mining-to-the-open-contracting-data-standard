```python
import json

import urllib.request

import pandas as pd

with open("CDMX.json",encoding="utf8") as f:
    data=json.load(f)
```

```python
Key planning documents are not provided
```

```python
a0=[]
for n in data["contratos"]:
    p=len(n["releases"])-1
    if (n["releases"][p]["tender"]["procurementMethod"] in ["selective","open","limited"]):
        if len(n["releases"][p]['planning']["documents"])>0:
            a0.append(0)
        else:
            a0.append(1)
```

```python
print(a0)
```

```python
Supplier address is PO box with no address or phone number
```

```python
a1=[]
for n in data["contratos"]:
    p=len(n["releases"])-1
    if (n["releases"][p]["tender"]["procurementMethod"] in ["selective","open","limited"]):
        e=0
        for m in n["releases"][p]["awards"]:
            for l in m["suppliers"]:
                if(l["contactPoint"]["telephone"]==None or l["address"]["streetAddress"]==None):
                    e=1
        a1.append(e)
```

```python
print(a1)
```

```python
Bidder that has never bid previously wins tender
```

```python
a2=[]
a4=[]
for n in data["contratos"]:
    p=len(n["releases"])-1
    for m in n["releases"][p]["tender"]["tenderers"]:
        a2.append(m["name"])
    if(n["releases"][p]["tender"]["procurementMethod"] in ["selective","open","limited"]):
        a3=0
        for s in n["releases"][p]["awards"]:
            for l in s["suppliers"]:
                if(a2.count(l["name"]) == 1):
                    a3=1
        a4.append(a3)
```

```python
print(a4)
```

```python
Perennial losing bidders give appearance of legitimate competition when they have no intention of actually winning
```

```python
a5=[]
a6=[]
a8=[]
for n in data["contratos"]:
    p=len(n["releases"])-1
    for s in n["releases"][p]["awards"]:
        for l in s["suppliers"]:
            a5.append(l["name"])            
    for o in n["releases"][p]["tender"]["tenderers"]:
        a6.append(o["name"]) 
    if(n["releases"][p]["tender"]["procurementMethod"] in ["selective","open","limited"]):
        a7=0
        for m in n["releases"][p]["tender"]["tenderers"]:
            if(a5.count(m["name"]) == 0 and a6.count(m["name"])>1):
                a7=1
        a8.append(a7)
```

```python
print(a8)
```

```python
Short notice to bidders to submit expression of interest or prepare bids
```

```python
a9=[]
for n in data["contratos"]:
    p=len(n["releases"])-1
    if (n["releases"][p]["tender"]["procurementMethod"] in ["selective","open","limited"]):
        if(n["releases"][p]["tender"]["numberOfTenderers"]==1):
            a9.append(1)
        else:
            a9.append(0)
```

```python
print(a9)
```

```python
More than 4,872,974 should be open
```

```python
a10=[]
for n in data["contratos"]:
    p=len(n["releases"])-1
    if (n["releases"][p]["tender"]["procurementMethod"] in ["selective","open","limited"]):
        if(n["releases"][p]["planning"]["budget"]["amount"]["amount"]>=4872974 and (n["releases"][p]["tender"]["procurementMethod"] in ["selective","limited"]) ):
            a10.append(1)
        else:
            a10.append(0)
```

```python
print(a10)
```

```python
t= pd.DataFrame({'missingPlanning':a0,'missingAdressPhone':a1,"never":a4,"loser":a8,"shortPeriod":a9,"highNotOpen":a10})
```
